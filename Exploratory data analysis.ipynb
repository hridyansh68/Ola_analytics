```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("login_hours.csv")
```

```python
df.head()
```

```python
df = df.sort_values(by=['car_number','active_date'],ascending=[True,True])
```

```python
df2 = df.head(20000)
```

df2

```python
countvalues = df2["car_number"].value_counts()
```

```python
sumseries = df2.groupby(["car_number"])["login_hours"].sum()
```

```python
countvalues = countvalues.sort_index()
```

```python
sumseries = sumseries.sort_index()
```

```python
sumseries
```

```python
df3 = pd.concat([countvalues,sumseries],axis=1)
```

```python
df3
```

```python
df3
```

```python
df3.values
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(df3.values)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(df3["login_hours"],df3["car_number"],c=kmeans.labels_)
```
